```python
#Import necessary libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from  sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
df= pd.read_csv('/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.sentiment.value_counts()
```

```python
df['Category']=df['sentiment'].apply(lambda x: 1 if x=='positive' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(df.review, df.Category, test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

**Using Random Forest**

```python
from sklearn.pipeline import Pipeline

clf= Pipeline([('vectorizer',CountVectorizer()),('random_forest',RandomForestClassifier(n_estimators=50, criterion='entropy'))])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred= clf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

**Using KNN**

```python
from sklearn.pipeline import Pipeline

clf2= Pipeline([('vectorizer',CountVectorizer()),('KNN',KNeighborsClassifier(n_neighbors=10, metric='euclidean'))])
```

```python
clf2.fit(X_train, y_train)
```

```python
y_pred= clf2.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

**Using Multinomial Naive Bayes**

```python
clf3= Pipeline([('vectorizer',CountVectorizer()),('nb',MultinomialNB())])
```

```python
clf3.fit(X_train, y_train)
```

```python
y_pred= clf3.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

We have achieved a good performance of above 80% using both the Multinomial Naive Bayes Model and the Random Forest model. KNN gives a poor performance as BoW uses large dimensional vectors and hence finding distance gets costly and hence impacts the performance of the model.
